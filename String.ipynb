```python
a='Today is 19-Jan-2026 MONDAY'
bc=0
vc=0
cc=0
for i in a:
  if i==" ":
    bc=bc+1
  elif i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
    vc=vc+1
  else:
    cc=cc+1
print(bc)
print(vc)
print(cc)
```

```python
a='Today is 20-Jan-2026 TUESDAY'
b=v=c=d=o=0
for i in a:
  if i.isspace():
    b=b+1
  elif i in "aeiouUEA":
    v=v+1
  elif i.isalpha():
    c=c+1
  elif i.isdigit():
    d=d+1
  else:
    o=o+1
print("space",b)
print("Vowels",v)
print("Consonants",c)
print("Digits",d)
print("Others",o)
```

Extracting ALL strings with a prefix

```python
codes=[
  "CRP_RICE_2026",
  "CRP_WHEAT_2025",
  "FRT_MANGO_2026",
  "CRP_MAIZE_2024"
]
filtered=[]
for crop_code in codes:
  if crop_code.startswith("CRP_"):
    filtered.append(crop_code)
print(filtered)
```

Extract and clean in one go

```python
clean_codes=[]
for code in codes:
  if code.startswith("CRP_"):
    clean_codes.append(code[4:])
print(clean_codes)
```

Split method

```python
for crop_code in codes:
  if crop_code.startswith("CRP_"):
    code,crop,year=crop_code.split('_')
    print(crop,year)
```

Mini Agriculture Data Cleaning Example                   
A consolidated example using multiple string operations.

```python
raw_record=" Rice,1200,Black Soil**"
clean_record=raw_record.strip()
clean_record1=clean_record.rstrip("*")
crop,yield_kg,soil=clean_record1.split(",")
crop=crop.title()

print(f"Crop:{crop}")
print(f"Yield:{yield_kg}kg")
print(f"Soil Type:{soil}")
```
